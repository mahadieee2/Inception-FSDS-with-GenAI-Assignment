# Assignment 1: Temperature Check

```python
""" 
 Ask the user for the current temperature

 1. "It's too cold!" if below 10°C
 2. "It's cool outside" if between 10°C and 25°C
 3. "It's hot outside!" if above 25°C

 """
   

temperature = int(input("Enter the current temperature in °C: "))


if temperature < 10:
    print("It's too cold!")
elif 10 <= temperature <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System

```python
"""
 Ask the user for their account balance and the amount they want to withdraw

 1. If the withdrawal amount is greater than the balance → print "Insufficient
 Balance"
 
 2. If the withdrawal amount is less than or equal to balance → print "Transaction
 Successful" and show the remaining balance.

 """

balance = float(input("Enter your account balance: "))

withdraw = float(input("Enter you  withdraw amount: "))



if withdraw > balance:
    print("Insufficient Balance")
else:
    remaining_balance = balance - withdraw
    print("Transaction Successful!")
    print("Remaining Balance:", remaining_balance)
```

# Assignment 3: Student Grade Calculator

```python
"""
Input Your marks (0–100) and print grade based on your marks
1. 90-100 = A+
2. 80-89 = A
3. 70-79 = B
4. 60-69 = C
5. Below 60 = Fail

"""

Number = float(input("Give Enter Your Number Here: "))




if Number >= 90 and Number <= 100:
    print("Your Grade is A+")
elif Number >= 80 and Number <= 89:
    print("Your Grade is A")
elif Number >= 70 and Number <= 79:
    print("Your Grade is B")
elif Number >= 60 and Number <= 69:
    print("Your Grade is C")

else:
    print("Your Grade is Fail")
```

# Assignment 4: Bus Ticket Discount

```python
"""

Ask the user for age and print ticket price:
● Below 5 → Free
● 5–18 → 50% Discount
● 60 or above → 30% Discount
● Others → Full Price

"""
# Ask user for there age.
age = int(input("Enter your Age Here: "))


if age < 5:
  print("Ticket Price: Free")
elif age >= 5 and age <= 18:
  print("Ticket Price: 50% Discount")
elif age >= 60:
  print("Ticket Price: 30% Discount")

else:
  print("Ticket Price: Full Price")
```

# Assignment 5: Restaurant Bill with Discount

```python
""" 

Take total bill amount from the user:

● If the bill is above 1000 → apply 10% discount
● Otherwise → no discount
Then print the final bill amount.

"""


bill = int(input("Enter Your Bill Here: "))

if bill > 1000:
    discount = bill * 0.10  
    final_bill = bill - discount
    print(final_bill)

else:
  print("No Discount")
  
```

# Assignment 8: Leap Year Checker

```python
""" 

Take a year as input and check:
● If it is divisible by 400 → Leap Year
● Else if divisible by 4 but not by 100 → Leap Year
● Otherwise → Not a Leap Year


"""


year = int(input("Enter a year: "))


if year % 400 == 0:
    print(year, "is a leap year.")
elif year % 4 == 0 and year % 100 != 0:
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

# Assignment 10: Electricity Bill Calculator

```python
""" 

Input number of units consumed and calculate bill:

● Up to 100 units → 5 BDT per unit
● 101–200 units → 7 BDT per unit
● Above 200 units → 10 BDT per unit
Print total payable amount.


"""



unit = int(input("Enter your unit here: "))


if unit <= 100:
  bill = unit * 5
  print(bill)

elif unit >= 101 and unit <= 200:
  bill = unit * 7
  print(bill)

else:
  actual_bill = unit * 10
  print("Total payable Amount", actual_bill)
```
